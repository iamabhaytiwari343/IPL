```python
#looking closely at the dataset we have last 2 columns empire 1 and umpire 2 we do not need 
#so after deleting those 2 columns and again saving the dataframe thus obtained in df
df = df.iloc[:, :14]
df

```

```python
city_counts = df['City'].value_counts()
cities = city_counts.index.tolist()
print(city_counts)
cities
```

```python
#plotting the matches played in different destinations till 2020
# Create a pie chart
plt.figure(figsize=(15, 15))
plt.pie(city_counts, labels=city_counts.index, autopct='%1.1f%%')
plt.title('Value Counts as a Percentage (Pie Chart)')

# Create a bar chart
plt.figure(figsize=(15, 15))
plt.bar(city_counts.index, city_counts)
plt.xticks(rotation=45)
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Value Counts (Bar Chart)')

# Show the plots
plt.show()


```
